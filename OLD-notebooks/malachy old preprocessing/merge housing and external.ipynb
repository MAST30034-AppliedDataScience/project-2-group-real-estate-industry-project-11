## Read Cinques file

```python
# donwload the housing data
housing_data = pd.read_csv("../../data/raw/historic/housing.csv")
housing_data.drop(columns="time stamp", inplace=True)
housing_data.rename(columns={"region": "suburbs"}, inplace=True)
housing_data.rename(columns={column: "housing: " + column for column in housing_data.columns if not column in ["year", "suburbs", "quarter"]}, inplace=True)
housing_data.head(5)
```

```python
# get the merged suburb data
merged_data = pd.read_csv("../../data/raw/historic/merged_as_suburbs.csv", index_col=0)
merged_data.head(5)
```

```python
# combine the frames
out = pd.merge(merged_data, housing_data, on=["suburbs", "year", "quarter"])
out.to_csv("../../data/raw/historic/final.csv", index=False)
```
