# Population Growth Analysis for Victorian SA2 Areas

This script analyzes population growth in Statistical Area Level 2 (SA2) regions of Victoria, Australia. It identifies and visualizes the top 10 fastest-growing areas.

## Data Loading and Preparation

The script starts by importing necessary libraries and loading the population data from a CSV file:


````python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

final_vic_population_filepath = '../data/curated/population/vic_population_df.csv'

final_vic_population_df = pd.read_csv(final_vic_population_filepath)

# Calculate population growth rate for each SA2 area
areas = final_vic_population_df['SA2 name'].unique()
growth_rates = []

for area in areas:
    area_data = final_vic_population_df[final_vic_population_df['SA2 name'] == area].sort_values('Year')
    if len(area_data) > 1:
        earliest_year = area_data['Year'].iloc[0]
        latest_year = area_data['Year'].iloc[-1]
        earliest_population = area_data['Total persons'].iloc[0]
        latest_population = area_data['Total persons'].iloc[-1]
        
        growth_rate = (latest_population - earliest_population) / earliest_population
        growth_rates.append((area, growth_rate, earliest_year, latest_year))

# Sort areas by growth rate and get top 10
top_10_areas = sorted(growth_rates, key=lambda x: x[1], reverse=True)[:10]

# Plotting
plt.figure(figsize=(14, 8))
for area, growth_rate, start_year, end_year in top_10_areas:
    area_data = final_vic_population_df[final_vic_population_df['SA2 name'] == area].sort_values('Year')
    plt.plot(area_data['Year'], area_data['Total persons'], marker='o', label=f"{area} ({start_year}-{end_year})")

plt.title('Population Growth for Top 10 Fastest Growing SA2 Areas')
plt.xlabel('Year')
plt.ylabel('Total Persons')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()

# Print growth rates
for area, growth_rate, start_year, end_year in top_10_areas:
    print(f"{area}: {growth_rate:.2%} growth from {start_year} to {end_year}")

# Save the plot
plt.savefig('top_10_fastest_growing_areas.png', dpi=300, bbox_inches='tight')

# Create the geolocations_notebook directory if it doesn't exist
output_dir = '../plots/geolocations_notebook'
os.makedirs(output_dir, exist_ok=True)

# Save the plot
output_file = os.path.join(output_dir, 'top_10_fastest_growing_areas.png')
plt.savefig(output_file, dpi=300, bbox_inches='tight')
````

````python
import folium
import geopandas as gpd

# Load the data
mega_filepath = '../data/raw/housing/merged_as_suburbs.csv'
mega_df = pd.read_csv(mega_filepath)

gdf_shapefile_joined_filepath = '../data/raw/location/gdf_shapefile_joined.shp'
gdf_shapefile_joined = gpd.read_file(gdf_shapefile_joined_filepath)

# Calculate median income for each suburb over the 4 years
median_income = mega_df.groupby('suburbs')['economic: median income'].median().reset_index()

# Merge the median income with the geospatial data
gdf_shapefile_joined = gdf_shapefile_joined.merge(median_income, left_on='suburbs', right_on='suburbs')

# Create a folium map
m = folium.Map(location=[-37.8136, 144.9631], zoom_start=10)  # Centered on Melbourne, adjust as needed

# Add the geospatial data to the map
folium.Choropleth(
    geo_data=gdf_shapefile_joined,
    name='choropleth',
    data=gdf_shapefile_joined,
    columns=['suburbs', 'economic: median income'],
    key_on='feature.properties.suburbs',
    fill_color='YlGnBu',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Median Income'
).add_to(m)

# Add GeoJson layer with customized style for clearer boundary lines
folium.GeoJson(
    gdf_shapefile_joined,
    style_function=lambda feature: {
        'color': 'black',
        'weight': 0.7,
        'fillOpacity': 0
    }
).add_to(m)

# Add layer control
folium.LayerControl().add_to(m)

# Save the map to an HTML file
m.save('income_distribution_map.html')

# Display the map in the notebook
m

# Create the geolocations_notebook directory if it doesn't exist
output_dir = '../plots/geolocations_notebook'
os.makedirs(output_dir, exist_ok=True)

# Save the map to an HTML file in the specified directory
output_file = os.path.join(output_dir, 'income_distribution_map.html')
m.save(output_file)
````

# School Distribution Choropleth Map for Victoria

This Python script creates an interactive choropleth map showing the distribution of schools across suburbs in Victoria, Australia.

## Dependencies

- os
- numpy
- pandas
- folium
- geopandas
- shapely

## Data Sources

The script uses two main data sources:
1. School data: `'../data/landing/school/schools.csv'`
2. Geospatial data: `'../data/raw/location/gdf_shapefile_joined.shp'`

## Process

1. **Data Loading**: 
   - Load the school data and geospatial data.

2. **Data Processing**:
   - Create a GeoDataFrame for schools using their coordinates.
   - Perform a spatial join to associate schools with suburbs.
   - Count the number of schools in each suburb.

3. **Map Creation**:
   - Initialize a Folium map centered on Victoria.

4. **Choropleth Generation**:
   - Merge school counts with geospatial data.
   - Create a choropleth layer using Folium's Choropleth function.
   - Customize the choropleth with color scheme, opacity, and legend.

5. **Map Display**:
   - Add the choropleth layer to the base map.
   - Display the interactive map.

## Key Features

- Interactive map centered on Victoria with zoom functionality.
- Choropleth overlay showing the number of schools in each suburb.
- Color-coded representation of school density across different areas.
- Legend indicating the number of schools corresponding to each color.

This visualization provides an intuitive way to understand the geographical distribution of schools across Victoria's suburbs. It can be particularly useful for identifying areas with high and low concentrations of schools, which could be valuable for educational planning, urban development, or demographic studies.

````python
import os
import numpy as np
import pandas as pd
import folium
import geopandas as gpd
from shapely.geometry import Point

school_filepath = '../data/landing/school/schools.csv'
school_df = pd.read_csv(school_filepath)

gdf_shapefile_joined_filepath = '../data/raw/location/gdf_shapefile_joined.shp'
gdf_shapefile_joined = gpd.read_file(gdf_shapefile_joined_filepath)

# Create a GeoDataFrame for the schools
school_gdf = gpd.GeoDataFrame(
    school_df,
    geometry=[Point(xy) for xy in zip(school_df['x'], school_df['y'])],
    crs=gdf_shapefile_joined.crs
)

# Perform a spatial join to associate schools with regions
school_gdf = gpd.sjoin(school_gdf, gdf_shapefile_joined, how='left', predicate='within')

# Count the number of schools in each region
school_counts = school_gdf.groupby('suburbs').size().reset_index(name='school_count')

# Merge the school counts with the geospatial data
gdf_shapefile_joined = gdf_shapefile_joined.merge(school_counts, on='suburbs', how='left')
gdf_shapefile_joined['school_count'] = gdf_shapefile_joined['school_count'].fillna(0)

# Create a base map centered around Victoria
victoria_map = folium.Map(location=[-37.4713, 144.7852], zoom_start=7)

# Add the geospatial data to the map
folium.Choropleth(
    geo_data=gdf_shapefile_joined,
    name='choropleth',
    data=gdf_shapefile_joined,
    columns=['suburbs', 'school_count'],
    key_on='feature.properties.suburbs',
    fill_color='YlOrRd',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Number of Schools'
).add_to(victoria_map)

# Display the map
victoria_map

# Create the geolocations_notebook directory if it doesn't exist
output_dir = '../plots/geolocations_notebook'
os.makedirs(output_dir, exist_ok=True)

# Save the map to an HTML file in the specified directory
output_file = os.path.join(output_dir, 'school_distribution_choropleth.html')
victoria_map.save(output_file)
````
